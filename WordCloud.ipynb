```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import FreqDist
from wordcloud import WordCloud
```

```python
nltk.download(["punkt","stopwords","movie_reviews","twitter_samples"])
```

```python
#Using isalpha method to include only the words that are made up of letters and to exclude punctuation marks

words = [w for w in nltk.corpus.movie_reviews.words() if w.isalpha()]
```

```python
#Using the common set of stop words available in English language

stopwords = nltk.corpus.stopwords.words("english")
```

```python
#Removing the stopwords by using lower() method to account for any discrepancies

words = [w for w in words if w.lower() not in stopwords]
```

```python
#Using FreqDist() method to count the frequency of each word in the corpus

fd = FreqDist(words)
fd
```

```python
fd.most_common(10)
```

```python
#Generating a wordcloud with the most frequent words

wordcloud = WordCloud(width = 800, height = 800, min_font_size = 10, background_color ='white').generate(str(words))
plt.figure(figsize = (10, 10), facecolor = 'red')
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
